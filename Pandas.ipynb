```python
import pandas as pd
# Pandas provide two data structures
# 1 - Series 2- Data-Frame

# Series
x = [3,4,5,6,7,8]
var = pd.Series(x)
print(var)
print(type(var))
print(var[2])
```

```python
# Changing Index Value
x = [3,4,5,6,7,8]
var =pd.Series(x,index=['a','b','c','d','e','f'])
print(var)
```

```python
# Changing Data Type
x = [3,4,5,6,7,8]
var =pd.Series(x,index=['a','b','c','d','e','f'],dtype="float")
print(var)
```

```python
# Name Parameter
x = [3,4,5,6,7,8]
var =pd.Series(x,index=['a','b','c','d','e','f'],dtype="float",name="Python")
print(var)
```
