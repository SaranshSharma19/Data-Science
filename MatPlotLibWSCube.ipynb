```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,color="r")
plt.title("Python",fontsize=15)
plt.xlabel("X-Axis",fontsize=15)
plt.ylabel("Y-Axis",fontsize=15)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,color="r")
plt.title("Python",fontsize=15)
plt.xlabel("X-Axis",fontsize=15)
plt.ylabel("Y-Axis",fontsize=15)
plt.fill_between(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,color="r")
plt.title("Python",fontsize=15)
plt.xlabel("X-Axis",fontsize=15)
plt.ylabel("Y-Axis",fontsize=15)
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,color="r")
plt.title("Python",fontsize=15)
plt.xlabel("X-Axis",fontsize=15)
plt.ylabel("Y-Axis",fontsize=15)
plt.fill_between(x,y,color="black")
plt.show()
```

```python
import numpy as np
x = [1,2,3,4,5]
x = np.array(x)
y = [1,2,3,4,5]
plt.plot(x,y,color="r")
plt.title("Python",fontsize=15)
plt.xlabel("X-Axis",fontsize=15)
plt.ylabel("Y-Axis",fontsize=15)
plt.fill_between(x,y,color="black",where=(x>=2) & (x<=4))
plt.show()
```

```python
import numpy as np
x = [1,2,3,4,5]
x = np.array(x)
y = [1,2,3,4,5]
plt.plot(x,y,color="r")
plt.title("Python",fontsize=15)
plt.xlabel("X-Axis",fontsize=15)
plt.ylabel("Y-Axis",fontsize=15)
plt.fill_between(x,y,color="black",where=(x>=2) & (x<=4),alpha=0.8)
plt.show()
```

# xticks(), yticks(), xlim(), ylim(), axis()

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
plt.plot(x,y)
plt.xticks(x,labels=["python","c","c++","c#","java"])
plt.yticks(y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
plt.plot(x,y)

plt.xlim(0,6)
plt.ylim(0,6)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]
plt.plot(x,y)
plt.axis([0,6,0,7])
plt.show()
```

# Working With Text In Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Programming Language",fontsize=15)
plt.xlabel("X-Axis",fontsize=15)
plt.ylabel("Y-Axis",fontsize=15)
plt.text(2.65,5.2,"Python",fontsize=15,style="italic",bbox={"facecolor":"y"})
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]
plt.plot(x,y)
plt.title("Programming Language",fontsize=15)
plt.xlabel("X-Axis",fontsize=15)
plt.ylabel("Y-Axis",fontsize=15)
plt.text(2.65,5.2,"Python",fontsize=15,style="italic",bbox={"facecolor":"y"})
plt.annotate("C++",xy=[4,2],xytext=(4,4),arrowprops=dict(facecolor="black",shrink=200))
plt.legend(["Python Data Science"], loc=)
plt.show()
```
